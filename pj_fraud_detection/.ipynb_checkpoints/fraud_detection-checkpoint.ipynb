```python
import pandas as pd

train_df = pd.read_csv(r"D:\Project\dataset\train_hsbc_df.csv")
test_df = pd.read_csv(r"D:\Project\dataset\test_hsbc_df.csv")
```

```python
print("Missing values in train_df:")
display(train_df.isnull().sum())
print("\nMissing values in test_df:")
display(test_df.isnull().sum())
```

```python
!pip install pandas scikit-learn xgboost joblib
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

categorical_features = ['age', 'gender', 'zipcodeOri', 'merchant', 'zipMerchant', 'category']
numerical_features = ['step', 'amount']

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='drop' # Drop the remaining columns (including 'customer' and 'fraud')
)

# Separate target variable before processing
train_target = train_df['fraud']
test_target = test_df['fraud']

train_df_processed = preprocessor.fit_transform(train_df.drop('fraud', axis=1))
test_df_processed = preprocessor.transform(test_df.drop('fraud', axis=1))


print("Processed train_df shape:", train_df_processed.shape)
print("Processed test_df shape:", test_df_processed.shape)
```

```python
!pip install pandas scikit-learn xgboost joblib matplotlib seaborn
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Distribution of target variable
print("Training dataset target distribution:")
display(train_target.value_counts())
print("\nTesting dataset target distribution:")
display(test_target.value_counts())

# 2. Visualize target variable distribution
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.countplot(x=train_target)
plt.title('Distribution of Fraud in Training Data')
plt.xlabel('Fraud')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
sns.countplot(x=test_target)
plt.title('Distribution of Fraud in Testing Data')
plt.xlabel('Fraud')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# 3. Distributions of numerical features
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.histplot(train_df['step'], kde=True)
plt.title('Distribution of Step in Training Data')

plt.subplot(1, 2, 2)
sns.histplot(train_df['amount'], kde=True, bins=50)
plt.title('Distribution of Amount in Training Data')

plt.tight_layout()
plt.show()
```

```python
# 5. Relationship between categorical features and target variable
categorical_features = ['age', 'gender', 'zipcodeOri', 'zipMerchant', 'category']

for feature in categorical_features:
    plt.figure(figsize=(10, 5))
    fraud_proportion = train_df.groupby(feature)['fraud'].mean().sort_values(ascending=False)
    fraud_proportion.plot(kind='bar')
    plt.title(f'Proportion of Fraud by {feature}')
    plt.xlabel(feature)
    plt.ylabel('Proportion of Fraud')
    plt.tight_layout()
    plt.show()

# Merchant has too many unique values to plot effectively, so we'll just print the top ones
print("\nProportion of Fraud by Merchant (Top 10):")
merchant_fraud_proportion = train_df.groupby('merchant')['fraud'].mean().sort_values(ascending=False)
display(merchant_fraud_proportion.head(10))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(train_df_processed, train_target, test_size=0.2, random_state=42)

print("Shape of X_train:", X_train.shape)
print("Shape of X_val:", X_val.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_val:", y_val.shape)
```

```python
import xgboost as xgb

# Instantiate the XGBoost model
# Using default parameters initially, consider 'scale_pos_weight' later if needed
xgboost_model = xgb.XGBClassifier(objective='binary:logistic', eval_metric='logloss', random_state=42)

# Train the model
xgboost_model.fit(X_train, y_train)

print("XGBoost model training complete.")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# Predict probabilities and labels on the validation set
y_pred_proba = xgboost_model.predict_proba(X_val)[:, 1] # Probabilities for the positive class
y_pred = xgboost_model.predict(X_val) # Predicted class labels

# Calculate evaluation metrics
accuracy = accuracy_score(y_val, y_pred)
precision = precision_score(y_val, y_pred)
recall = recall_score(y_val, y_pred)
f1 = f1_score(y_val, y_pred)
roc_auc = roc_auc_score(y_val, y_pred_proba)

# Print the metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"ROC AUC: {roc_auc:.4f}")
```

```python
# Predict probabilities for the positive class (fraud = 1) on the processed test dataset
test_predictions_proba = xgboost_model.predict_proba(test_df_processed)[:, 1]

print("Test predictions probabilities generated.")
```

```python
import pandas as pd

# Create a DataFrame with the predictions
submission_df = pd.DataFrame({
    'customer': test_df['customer'], # Include the customer column
    'fraud': test_predictions_proba # Predicted probabilities
})

# Save the DataFrame to a CSV file without the index
submission_df.to_csv('submission.csv', index=False)

print("Submission file 'submission.csv' created successfully with customer column.")
```
